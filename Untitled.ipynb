```python
def print_row_vertically(df, row_index=0):
    """
    Prints a single row of a DataFrame in a vertical format.
    By default, it prints the first row (index 0).
    """
    # 1. Catch errors if the row index doesn't exist
    if row_index >= len(df) or row_index < 0:
        print(f"Error: Row index {row_index} is out of bounds.")
        return

    # 2. Extract the requested row (this turns it into a pandas Series)
    row = df.iloc[row_index]
    
    # 3. Find the length of the longest column name for perfect alignment
    max_col_length = max(len(str(col)) for col in df.columns)
    
    # 4. Loop through each column and value, printing them in your format
    for col_name, value in row.items():
        # The :<{max_col_length} part left-aligns the column name 
        # and pads it with spaces so the values start at the exact same spot
        print(f"--| {str(col_name):<{max_col_length}}      {value}")
```

```python
import pandas as pd
import plotly.express as px
import webbrowser
import os
import pandas as pd
import plotly.graph_objects as go
from plotly.subplots import make_subplots
from pptx import Presentation  # Added for PowerPoint
from pptx.util import Inches     # Added for sizing images
from bidi.algorithm import get_display
import pandas as pd
import numpy as np
import re

from sentence_transformers import SentenceTransformer
import hdbscan
team_to_filter = 4
def is_valid(text):
    text = str(text).strip()
    
    # remove very short
    if len(text) < 5:
        return False
    
    # remove dots / punctuation-only
    if re.fullmatch(r"[.\-_,\s]+", text):
        return False
    
    # remove repeated names (like ארבל)
    if len(text.split()) == 1:
        return False
    
    return True

# 1. Identify all numeric columns
df = pd.read_excel(r"file.xlsx")
#df = df[df['צוות:'] == team_to_filter]
# 2. Ensure the 'Timestamp' column is in a proper datetime format
df['Timestamp'] = pd.to_datetime(df['Timestamp'])

# 3. Identify all numeric columns automatically
non_numeric_cols = df.select_dtypes(include=['object'])

df_long = non_numeric_cols.melt(value_name="comment").dropna()

# Keep only text
df_long = df_long[df_long["comment"].astype(str).str.strip() != ""]
df_long = df_long[df_long["comment"].apply(is_valid)]
# Create pattern column
# Count occurrences
counts = df_long["comment"].value_counts()

# Keep only values that appear ONCE
df_long = df_long[df_long["comment"].map(counts) < 3]

# =========================
# 3. Hebrew text cleaning
# =========================


def clean_hebrew(text):
    text = str(text)

    # remove punctuation
    text = re.sub(r"[^\w\s\u0590-\u05FF]", " ", text)

    # normalize Hebrew final letters
    replacements = {
        "ך": "כ",
        "ם": "מ",
        "ן": "נ",
        "ף": "פ",
        "ץ": "צ"
    }
    for k, v in replacements.items():
        text = text.replace(k, v)

    # remove extra spaces
    text = re.sub(r"\s+", " ", text).strip()

    return text

df_long["clean_comment"] = df_long["comment"].apply(clean_hebrew)

# =========================
# 4. Generate embeddings
# =========================
model = SentenceTransformer("paraphrase-multilingual-MiniLM-L12-v2")

embeddings = model.encode(
    df_long["clean_comment"].tolist(),
    show_progress_bar=True
)

# =========================
# 5. Cluster (HDBSCAN)
# =========================
clusterer = hdbscan.HDBSCAN(
    min_cluster_size=5,        # smaller = more clusters
    min_samples=2,             # more sensitive
    cluster_selection_epsilon=0.1  # splits clusters more
)

labels = clusterer.fit_predict(embeddings)

df_long["cluster"] = labels

# =========================
# 6. Inspect clusters
# =========================
def print_clusters(df, n_samples=5):
    clusters = sorted(df["cluster"].unique())

    for c in clusters:
        print("="*50)
        print(f"Cluster {c}")
        
        sample = df[df["cluster"] == c]["comment"].head(n_samples)
        for s in sample:
            print("-", s)

print_clusters(df_long)
```

```python
import pandas as pd
import plotly.express as px
import webbrowser
import os
import plotly.graph_objects as go
from plotly.subplots import make_subplots
from pptx import Presentation  # Added for PowerPoint
from pptx.util import Inches     # Added for sizing images
from bidi.algorithm import get_display

team_to_filter = 3

# 1. Read data
df = pd.read_excel(r"file.xlsx")
# df = df[df['צוות:'] == team_to_filter]

# 2. Ensure the 'Timestamp' column is in a proper datetime format
df['Timestamp'] = pd.to_datetime(df['Timestamp'])

# (Optional safety check): If 'week' column doesn't exist, create it from Timestamp
if 'week' not in df.columns:
    df['week'] = df['Timestamp'].dt.isocalendar().week

# 3. Identify all numeric columns automatically
numeric_cols = df.select_dtypes(include=['number']).columns.tolist()
z_score_value = 1.645

# --- Calculate absolute Z-scores and > 2 outlier flags ---
agg_dict = {}
for col in numeric_cols:
    m = df[col].mean()
    s = df[col].std()
    
    # Calculate raw Z-score
    z_raw = (df[col] - m) / s
    df[f'{col}_zscore_raw'] = z_raw.fillna(0)
    df[f'{col}_zscore_abs'] = z_raw.abs().fillna(0)
    
    # Flag positive and negative outliers based on 1.645
    df[f'{col}_pos_outlier'] = (df[f'{col}_zscore_raw'] > z_score_value).astype(int)
    df[f'{col}_neg_outlier'] = (df[f'{col}_zscore_raw'] < -z_score_value).astype(int)
    
    agg_dict[col] = ['mean', 'std']
    agg_dict[f'{col}_zscore_abs'] = ['mean'] 
    agg_dict[f'{col}_pos_outlier'] = ['sum']
    agg_dict[f'{col}_neg_outlier'] = ['sum']

# 4. Group the data into WEEKLY buckets
grouped = df.groupby(pd.Grouper(key='week')).agg(agg_dict).reset_index()

# Safely extract the week column for the X-axis
x_axis_weeks = grouped['week'] if 'week' in grouped.columns else grouped[('week', '')]

# Create a string to hold all our HTML
html_content = "<html><head><title>Survey Report</title></head><body>"
html_content += "<h1 style='text-align:center;'>Survey Analysis Results</h1>"

for col in numeric_cols:
    flag = 0
    # Count the occurrences of each score (1, 2, 3, etc.)
    counts = df[col].value_counts().reset_index()
    counts.columns = ['Score', 'Count']
    counts = counts.dropna(subset=['Score'])
    try:
        counts['Score'] = counts['Score'].astype(float).astype(int).astype(str).str.strip()
    except:
        flag = 1
        pass
        
    # Skip columns that are empty or have only NaN values
    if counts.empty:
        continue

    fig = go.Figure()

    # Calculate padding for Y-axis
    y_values = pd.concat([grouped[col]['mean'], grouped[col]['std'], grouped[f'{col}_zscore_abs']['mean']])
    y_min, y_max = y_values.min(), df[col].max() # updated to check raw max for boxplot fit
    padding = (y_max - y_min) * 0.9 if y_max != y_min else 5
    
    # --- Trace 1: The Box Plot (NEW) ---
    # This plots the distribution and keeps the raw values visible as dots
    fig.add_trace(go.Box(
        x=df['week'], 
        y=df[col],
        name='Distribution',
        boxpoints='all',      # This is the key to keeping values visible!
        jitter=0.3,           # Spreads the dots out so they don't overlap entirely
        pointpos=-1.8,        # Shifts the dots slightly to the left of the box
        marker=dict(color='rgb(100, 149, 237)', size=6),
        line=dict(color='rgba(100, 149, 237, 0.8)'),
        fillcolor='rgba(100, 149, 237, 0.3)'
    ))
    
    # --- Trace 2: Average Score ---
    avg_values = grouped[col]['mean'].round(2)
    fig.add_trace(go.Scatter(
        x=x_axis_weeks, 
        y=grouped[col]['mean'],
        mode='markers+text',
        name='Average',
        text=grouped[col]['mean'].round(2),
        textposition="top center",
        textfont=dict(size=38, family="Heebo", color="black"), 
        marker=dict(size=36, color='blue'),
        line=dict(width=6, color='blue'),
        cliponaxis=False
    ))
    
    avg_z = grouped[f'{col}_zscore_abs']['mean'].round(2)
    pos_counts = grouped[f'{col}_pos_outlier']['sum'].astype(int)
    neg_counts = grouped[f'{col}_neg_outlier']['sum'].astype(int)
    z_labels = [f"סטיית תקן - {z}<br>(חריגים לחיוב: {p})<br>(חריגים לשלילה: {n})" for z, p, n in zip(avg_z, pos_counts, neg_counts)]
    
    # --- Trace 3: Std Dev (Position: BOTTOM) ---
    fig.add_trace(go.Scatter(
        x=x_axis_weeks, 
        y=grouped[col]['std'],
        mode='markers+text',
        name='Std Dev',
        text=z_labels,
        textposition="bottom center",
        textfont=dict(size=32, family="Heebo", color="black"), 
        marker=dict(size=36, color='red'),
        line=dict(width=6, color='red'),
        cliponaxis=False
    ))

    # --- Layout Fixes for Clipping ---
    fig.update_layout(
        height=850, # Increased height to give labels more vertical breathing room
        margin=dict(l=120, r=120, t=150, b=120), # Increased margins so labels don't cut off at edges
        hovermode="x unified",
        showlegend=True,
        legend=dict(font=dict(size=22), orientation="h", yanchor="bottom", y=1.02, xanchor="right", x=1),
        template='plotly_white', 
        font=dict(family="Arial, sans-serif", size=18, color="#333333"), 
        title=dict(text=f"{col}\u200F", font=dict(size=28, color="#111111"), x=0.5, xanchor='center'),
        yaxis=dict(range=[y_min - padding, y_max + padding])
    )
    
    # 4. Style the X-Axis (Bottom)
    fig.update_xaxes(
        showline=True, linewidth=2, linecolor='black', 
        showgrid=False,                                
        title_font=dict(size=20, color="black"),
        tickfont=dict(size=16, color="black"),
        dtick=1
    )
    
    # 5. Style the Y-Axis (Left)
    fig.update_yaxes(
        showline=True, linewidth=2, linecolor='black', 
        showgrid=True, gridwidth=1, gridcolor='#E5E5E5',
        title_font=dict(size=20, color="black"),
        tickfont=dict(size=16, color="black"),
        zeroline=False, zerolinewidth=2, zerolinecolor='black' 
    )
    
    # 6. Make the lines thicker
    fig.update_traces(line=dict(width=4), selector=dict(type='scatter'))
    
    # 7. Add figure to HTML
    html_content += fig.to_html(full_html=False, include_plotlyjs='cdn')

html_content += "</body></html>"

# Save the report
report_name = "Survey_Report.html"
with open(report_name, "w", encoding="utf-8") as f:
    f.write(html_content)

# Open the report automatically in the browser
webbrowser.open('file://' + os.path.realpath(report_name))
```
